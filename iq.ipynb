```python
# 1 to 100 in reverse

def reverseorder(n):

  if (n<=0):
    pass
  else:
    print(n, end=' ')
    reverseorder(n-1)

n=100;
reverseorder(n)
```

```python
# factorial of 29

num = 29
fact = 1

for i in range(1,num-1):
  fact = fact*i

  print(fact)
```
