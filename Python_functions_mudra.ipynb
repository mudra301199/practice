```python
def fib(num):
  if num <= 1:
    return num
  else:
    return(fib(num-1) + fib(num-2))
n = int(input('Enter a number: '))
if n <= 0:
  print("Plese enter a positive integer")
else:
  print("Fibonacci sequence:")
for i in range(n):
  print (fib(i))
```

```python
def add(x, y, *z):
  s = x + y
  for n in z:
    s += n
  else:
    return s
```

```python
def sub(x, y, *z):
  s = x - y
  for n in z:
    s -= n
  else:
    return s
```

```python
def mul(x, y, *z):
  s = x * y
  for n in z:
    s *= n
  else:
    return s
```
